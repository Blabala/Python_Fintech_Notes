&emsp;&emsp;自然界和社会上发生的现象是多种多样的，有一类现象，在一定条件下必然发生，例如，向上抛一个石子必然下落，同性电荷必相互排斥，等等。这类现象称为**确定性现象**。在自然界和社会上还存在着另一类现象，例如，在相同条件下抛同一枚硬币，其结果可能是正面朝上，也可能是反面朝上。人们在经过长期时间并深入研究之后，发现这类现象在大量重复试验或观察下，它的结果却呈现出某种规律性，这种在大量重复试验或观察中所呈现的固有规律性，就是我们以后所说的**统计规律性**。
  
&emsp;&emsp;这种在个别试验中其结果呈现出不确定性，在大量重复试验中其结果又具有统计规律性的现象，我们称之为**随机现象**。概率论与数理统计是研究和揭示随机现象统计规律性的一门数学学科。

## $\S1.1$ 随机试验
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;我们遇到过各种试验，例如抛一枚硬币，观察正面$H$、反面$T$出现的情况。在这个试验中有两种可能结果，出现$H$或者出现$T$，但在抛掷之前不能出现$H$还是出现$T$，这个试验可以在相同的条件下重复地进行。概括这一类试验具有以下几个特点：
+ 可以在相同的条件下重复地进行；
+ 每次试验的可能结果不止一个，并且能事先明确试验的所有可能的结果；
+ 进行一次试验之前不能确定哪一个结果会出现。

在概率论中，我们将具有上述三个特点的试验称为**随机试验**。

## $\S1.2$ 样本空间、随机事件
### （一）样本空间
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;对于随机试验，尽管我们不能预知试验的结果，但试验的所有可能结果组成的集合是已知的。我们将随机试验$E$的所有结果组成的集合成为$E$的**样本空间**,记为$S$。样本空间的元素，即$E$的每个结果，称为**样本点**。  
例如，1中的试验$E_{抛硬币}$的样本空间$S_{抛硬币}$：$$S_{抛硬币}=\{H、T\}$$其中样本点为正面$H$,反面$T$。
### （二）随机事件
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;一般，我们称试验$E$的样本空间$S$的子集为$E$的**随机事件**，简称为**事件**。在每次试验中，当且仅当这一子集中的一个样本点出现时，称为一**事件发生**。特别的，由一个样本点组成的单点集，称为**基本事件**，例如抛硬币中的两个基本事件为$\{T\}和\{H\}$。  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;样本空间$S$包含所有的样本点，它是$S$自身的子集，在每次试验中它总是发生的，$S$称为**必然事件**。空集$\emptyset$不包含任何样本点，它作为样本空间的子集，他在每次试验中都不发生，$\emptyset$称为**不可能事件**。
### （三）事件间的关系与事件的运算
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;设试验$E$的样本空间为$S$,而$A、B、A_k(k=1,2,...)$是$S$的子集。
+ 若$A\subset B$，则称事件$B$包含事件$A$，这指的是事件$A$发生必导致事件$B$发生。若$A\subset B且B\subset A$,则称事件$A$与时间$B$**相等**。
+ 事件$A\cup B={x|x\in A 或x\in B}$称为事件$A$与事件$B$的**和事件**，当且仅当$A$、$B$中至少有一个发生时，事件$A\cup B$发生。  
类似地，称$\bigcup\limits_{k=1}^{n}A_k$为$n$个事件$A_1,A_2,...,A_n$的和事件；
+ 事件$A\cap B=\{x|x\in A 且x\in B\}$称为事件$A$与事件$B$的**积事件**。当且仅当$A、B$同时发生时，事件$A\cap B$发生，$A\cap B$也记作$AB$。
类似地，称$\bigcap\limits_{k=1}^{n}A_k$为$n$个事件$A_1,A_2,...,A_n$的积事件；
+ 事件$A-B={x|x\in A且x\in B}$称为事件$A$和事件$B$的**差事件**。当且仅当$A$发生、$B$不发生事件$A-B$发生。
+ 若$A\cap B=\emptyset$，则称事件$A$与$B$是**互不相容**的，或**互斥**的。这指的是事件$A$与事件$B$不能同时发生。基本事件是两两互不相容的。
+ 若$A\cup B=S且A\cap B=\emptyset$，则称事件$A$与事件$B$互为**逆事件**，又称事件$A$与事件$B$互为**对立事件**。这指的是对每次试验而言，事件$A$、$B$中必有一个发生，且仅有一个发生。$A$的对立事件记为$\overline A$。$\overline A=S-A$。  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;在进行事件运算时，经常要用到下述定律，设$A,B,C$为事件，则有  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;1. 交换律：$A\cup B=B\cup A;A\cap B=B\cap A$.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2. 结合律：$A\cup(B\cup C)=(A\cup B)\cup C$&nbsp;;&nbsp;$A\cap (B\cap C)=(A\cap B)\cap C$.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;3. 分配律：$A\cup(B\cap C)=(A\cup B)\cap(A\cup C)$&nbsp;;&nbsp;$A\cap(B\cup C)=(A\cap B)\cup(A\cap C)$.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;4. 德摩根律：$\overline{A\cup B}=\overline A\cap \overline B$&nbsp;;&nbsp;$\overline{A\cap B}=\overline A\cup \overline B$

#### 例题：</font>
设将一枚硬币抛掷三次为$E$事件，其中$A_1$：“第一次出现的是$H$”,即

$$A_1=\{HHH,HHT,HTH,HTT\}$$事件$A_2$:“三次出现同一面”，即

$$A_2=\{HHH,TTT\}$$则有：$$A_1\cup A_2=\{HHH、HHT、HTH、HTT、TTT\}$$

$$A_1\cap A_2=\{HHH\}$$

$$A_2-A_1=\{TTT\}$$

$$\overline{A_1\cup A_2}=\{THT、TTH、THH\}$$

## $\S1.3$ 频率与概率
### （一）频率
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**定义**：在相同的条件下，进行了n次试验，在这n次试验中，事件A发生的次数 ，称为事件A发生的频数，比值$n_A$/$n$称为事件A发生的频率，并记成 $f_n(A)$。则有 ：

$$f_n(A)=\frac{n_A}{n}$$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;有定义，可见频率具有下述基本性质：
+ $0\le f_(A)\le 1$;

+ $f_n(S)=1$;

+ 若$A_1,A_2,...,A_k$是两两互不相容的事件，则

$$f_n(A_1\cup A_2\cup ...\cup A_k)=f_n(A_1)+f_n(A_2)+...+f_n(A_k)$$。

#### 例题：</font>
你从一副扑克牌中抽一张牌，抽到A的频率有多大？
$$P(抽到A)=\frac{一副牌中A的数量}{一副牌中所有牌的数量}=\frac{4}{52}=0.769$$

#### 例题：</font>
同时抛弃5枚硬币，如果正面朝上少于3枚，则输掉8元，否则就赢8元。如果手中有1000元作为赌资，请问赌博10000次后可能会是什么情况呢？

```python
import numpy as np#导入画图模块
from matplotlib.pyplot import plot,show#导入matplotlib模块的画图和显示函数
money=np.zeros(10000)#先将赌博次数的每一次的money初始化为0money[0]=1000#表示刚开始有1000元现金
result=np.random.binomial(5,0.5,size=len(money))#二项分布的10000次成功的次数
#遍历10000数，更新每一次的moeny值存储
for i in range(1,len(money)):#从1到10000进行遍历    
    if result[i]<3:       
        money[i]=money[i-1]-8    
    else:       
        money[i]=money[i-1]+8
print(result.min(),result.max())#打印5次抛硬币最多出现正面，和最少出现正面的次数
print(money.max(),money.min())#打印可能出现钱最多，最少的情况
plot(np.arange(len(money)),money)#画图show()#显示输出结果：
```

### （二）概率
**定义**设$E$是随机试验，$S$是它的样本空间，对于$E$的每一个事件$A$赋予一个实数，记为$P(A)$,称为事件$A$的**概率**，则有

$$ P(A)=\lim_{n_t}\frac{n_A}{n_t}$$  
如果集合函数$P$满足下列条件：
+ **非负性**：对于每一个事件$A$,有$P(A)\ge0$;
+ **规范性**：对于必然事件$S$,有$P(S)=1$;
+ **可列可加性**： 设$A_1,A_2,...$是两两互不相容的事件，即对于$A_iA_j=\emptyset,i\ne j,i,j=1,2,...$，有

$$P(A_1\cup A_2\cup...)=P(A_1)+P(A_2)+...$$
由概率的定义，可以推出概率的一些重要性质：
#### 性质1** $P(\emptyset=0)$**
证： 令$A_n=\emptyset(n=1,2,...)$,则$\bigcup\limits_{n=1}^{\infty}A_n=\emptyset$，且$A_iA_j=\emptyset,i\ne j,i,j=1,2,...$,由概率的可列可加性得：

$$P(\emptyset)=p(\bigcup\limits_{n=1}^{\infty}A_n)=\sum_{i=1}^{n} P(A_n)=\sum_{i=1}^{n} P(\emptyset)$$
由概率的非负性知,$P(\emptyset)\ge 0$，故由上式知$P(\emptyset)=0$
#### 性质2 有限可加性：若$A_1,A_2,...,A_n$是两个互不相容的事件，则有

$$P(A_1\cup A_2\cup...\cup A_n)=P(A_1)+P(A_2)+...+P(A_n)$$
证：令$A_{n+1}=A_{n+2}=...=\emptyset$，即有$A_iA_j=\emptyset,i\ne j,i,j=1,2,...$,由可列可加性有

$P(A_1\cup A_2\cup...\cup A_n)\\=P(\bigcup\limits_{n=1}^{\infty}A_n)=\sum_{i=1}^{n}P(A_i)\\=\sum_{i=1}^{n}P(A_i)+0=P(A_1)+P(A_2)+...+P(A_n)$
#### 性质3.设A,B是两个事件，若$A\in B$ ，则有

$$P(B-A)=P(A)-P(B)\\P(B)\ge P(A)$$
证：由$A\in B$知$B=A\cup(B-A)$,且$A(B-A)=\emptyset$,再由有限可加性得，

$$P(B)=P(A)+P(B-A)$$,在由概率的非负性，$P(B-A)\ge 0$

$$P(B)\ge P(A)$$。

#### 性质4.对于任一事件A，$$P(A)≤1$$
证：因$A\in S$，有上面的性质得

$$P(A)\le P(S)=1$$
#### 性质5.对于任一事件A，有 $$P(\bar{A})=1-P(A)$$
证 因$A\cup\bar{A}=S$,且$A\bar{A}=\emptyset$，得

$$1=P(S)=P(A\cup\bar{A})=P(A)+P(\bar{A})$$
#### 性质6.对于任意两事件A,B有

$$P(A\cup B)=P(A)+P(B)-P(AB)$$
证：因$A\cup B=A\cup(B-AB)$，且$A(B-AB)=\emptyset,AB\in B$,得

$P(A\cup B)=P(A)+P(B-AB)\\=P(A)+P(B)-P(AB)$
 
推广到多个事件的情况，例如，设$A_1,A_2,A_3,...,A_n$为n个事件，则可以用归纳法有
  
$$  P(A_1\cup A_2\cup …\cup A_n ）=\sum_{i=1}^{n}P(A_i) -\sum_{1\le i<j\le n}P(A_iA_j)\\+\sum_{1\le i<j<k\le n}P(A_iA_j) +…+(-1)^(n-1)P(A_1A_2...A_n)$$


## $\S1.4$ 等可能概型（古典概型）
**定义**
具有以下两个特点的试验称为等可能概型：  
1.试验的样本空间只包含有限个元素  
2.试验中每个基本事件发生的可能性相同  
具有以上两个特点的试验是大量存在的，这种试验称为**等可能概型**，它在概率论发展初期曾是主要的研究对象，所以也称为**古典概型**。  
下面我们来讨论等可能概型中事件概率的计算公式。
设试验的基本空间为$S=\{e_1,e_2,...,e_n\}$，由于在试验中每个基本事件发生的可能性相同，即有
$$P(\{e_1\}=P(\{e_2\}=...P(\{e_n\})$$
又由于基本事件是两两互不相容的，于是

$$1=P(S)=P(\{e_1\}+P(\{e_2\}+...P(\{e_n\})\\=nP(\{e_i\})\\P(\{e_i=\frac{1}{n}\}),i=1,2,...,n$$
即

$$P(A)=\sum_{i=1}^{k}P(\{e_i\})=\frac{k}{n}=\frac{A包含的基本事件数}{S中基本事件的总数}$$

#### 例题：</font>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;将一枚硬币抛掷三次（1）设事件$A_1$为“恰有一次出现正面”，求$P(A_1)$;(2)设事件$A_2$为“至少有一次出现正面”，求$P（A_2)$

(1)$$P(A_1)=\frac{3}{8}$$
(2)$$P(A_2)=1-\overline P(A_2)=1-\frac{1}{8}=\frac{7}{8}$$
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;人们在长期的实践中总结得到**“概率很小的事件在一次试验中实际上几乎是不发生的”**，称之为**实际推断原理**。

## $\S1.5$ 条件概率
### （一）条件概率
条件概率是概率论中的一个重要而实用的概念，所考虑的是事件A已发生的条件下事件B发生的概率。  
**定义** 设A，B是两个事件，且$P(A)>0$，称

$$P(B|A)=\frac{P(AB)}{P(A)}$$
为在事件A发生的条件下事件B发生的**条件概率**。
由此定义可以得到条件概率P(•｜A)的性质：
+ 非负性:$P(B｜A)≥0$;
+ 规范性：对于必然事件S，有$P(S｜A)=1$
+ 可列可加性：设$B_1,B_2…$是两两互不相容的事件，则有

$$P(\bigcup\limits_{i=1}^{\infty}B_i|A)=\sum_{i=1}^{\infty}P(B_i|A)$$
#### 例题：</font>
甲袋中有5只白球, 7 只红球;乙袋中有4只白球, 2只红球.从两个袋子中任取一袋, 然后从所取到的袋子中任取一球,求取到的球是白球的概率.   
设事件A表示“取到的是甲袋”, 则 表示“取到的是乙袋”,事件B表示“最后取到的是白球”.
根据题意:  $$P(B|A)=\frac{5}{12}\\P(B|\bar{A})=\frac{2}{3}\\P(A)=\frac{1}{2}\\P(B)=P(B|A)P(A)+P(B|\bar{A})P(\bar{A})\\=\frac{5}{12}*\frac{2}{3}*(1-\frac{1}{2})=\frac{13}{24}$$ 
### （二）乘法原理
由条件概率的定义，立即可得到下述定理:
**乘法原理**：设$P(A)>0$，则有$$P(AB)=P(B｜A)P(A)$$
### （三）全概率公式和贝叶斯公式
**定义** ：设S为试验E的样本空间,$B_1,B_2,...,B_n$为E的一组事件，若 

(1)$B_iB_j=\emptyset,i\ne j,i,j=1,2,...,n$;

(2)$B_1\cup B_2\cup …\cup B_n=S$;  
则称$B_1,B_2,...,B_n$为样本空间$S$的一个**划分**。

#### 例题：</font>
 据美国的一份资料报导，在美国总的来说患肺癌的概率约为$0.1\%$，在人群中有$20\%$是吸烟者，他们患肺癌的概率约为$0.4\%$，求不吸烟者患肺癌的概率是多少？
 解：以C事件“患肺癌”，以A记事件“吸烟”，按题意$P(C)=0.001,P(A)=0.20,P(C|A)=0.004$,需要求条件概率$P(C|\bar{A})$，由全概率公式有$$P(C)=P(C|A)P(A)+P(c|\bar{A})$$将数据代入，得$$0.001=0.004*0.20+P(C|\bar{A})P(A)\\=0.004*0.20+P(C|\bar(A))*0.80\\P(C|\bar{A})=0.00025$$
 

#### 例题：</font>
三门问题（Monty Hall problem）亦称为蒙提霍尔问题、蒙特霍问题或蒙提霍尔悖论，大致出自美国的电视游戏节目Let's Make a Deal。问题名字来自该节目的主持人蒙提·霍尔（Monty Hall）。参赛者会看见三扇关闭了的门，其中一扇的后面有一辆汽车，选中后面有车的那扇门可赢得该汽车，另外两扇门后面则各藏有一只山羊。当参赛者选定了一扇门，但未去开启它的时候，节目主持人开启剩下两扇门的其中一扇，露出其中一只山羊。主持人其后会问参赛者要不要换另一扇仍然关上的门。问题是：换另一扇门会否增加参赛者赢得汽车的机率？ 

```python
from random import choice

def stay():
    doors = ['car','goat','goat']       #设置三扇门，其中两扇门后面是山羊，一扇门后是汽车
    choose = choice(doors)              #随机选择一扇门
    if choose == 'car':                 #不换门则直接判断
        return 'win'
    else:
        return 'lose'

def switch():
    doors = ['car', 'goat', 'goat']
    choose = choice(doors)
    doors.remove(choose)               #选择的门放一边
    doors.remove('goat')               #剩下的门，去掉一只山羊
    if doors == ['car']:               #换门
        return 'win'
    else:
        return 'lose'



if __name__ == '__main__':
    total = 100000
    count_switch = 0
    win_switch = 0
    count_stay = 0
    win_stay = 0
    for i in range(total):
        choose = choice([1,2])          #随机选择换门还是不换门
        if choose == 1:
            count_switch += 1
            if switch() == 'win':
                win_switch += 1
        else:
            count_stay += 1
            if stay() == 'win':
                win_stay += 1

    print('换门次数:',count_switch)
    print('换门后得到汽车:',win_switch,'%.2f%%'%(100*win_switch/count_switch))
    print('不换门:',count_stay)
    print('不换门后得到汽车:',win_stay,'%.2f%%'%(100*win_stay/count_stay)) 
```

## $\S1.6$ 独立性
**定义** ：设A、B是两事件，如果满足如下等式，则称事件A、B**相互独立**，简称A、B**独立**。容易得到，若$P(A)>0,P(B)>0$，则A，B互相独立与A，B互不相容不能同时成立。  
$$P(AB)=P(A)P(B)$$


**定理一**：
设A,B是两事件，且P(A)>0，若A,B相互独立，则P(B｜A)=P(B),反之亦然。  

**定理二**：
若事件A与B相互独立，则下列各对事件也相互独立：A与$\bar{B}$ , $\bar{A}$与B, $\bar{A}$与$\bar{B}$    
**证** 因为$A=A(B\cup \bar{B})=AB\cup A\bar{B}$,得到$$P(A)=P(AB\cup A\bar{B})=P(AB)+P(A\bar{B})\\=P(A)P(B)+P(A\bar{B})\\P(A\bar{B})=P(A)[1-P(B)]=P(A)P(\bar{B})$$

**定义**设A,B,C是三个事件，如果满足等式如下，则称事件A,B,C相互独立。
$$P(AB)=P(A)P(B),\\P(BC)=P(B)P(C)\\P(AC)=P(A)P(C),\\P(ABC)=P(A)P(B)P(C)$$

**推论**：  

1.若事件$A_1,A_2,...,A_n$(n≥2)相互独立，则其中任意k(2≤k≤n)个事件也是相互独立；

2.若n个事件 $A_1,A_2,...,A_n$(n≥2)相互独立，则将 … 中任意多个事件换成它们的对立事件，所得的n各事件仍相互独立

## $\S2.1$ 离散型随机变量和连续型随机变量的分布规律
#### 离散型随机变量的分布规律
设离散型随机变量$X$的可能取值为$X_k$$(k=1,2,\ldots)$且取各个值的概率,即事件$(X=x_k)$的概率为:
$$P(X=x_k)=p_k，k=1,2,…，$$  
显然分布律应满足下列条件:  
$1.p_k\geq0，k=1,2,…,n$

$2.\sum_{k=1}^\infty{p_k}=1$

#### 连续型随机变量的分布规律
设$F(x)$是随机变量$X$的分布函数，若存在非负函数$f(x)$，对任意实数，有$$F(x)=\int_{-\infty}^x{f(x)}{\rm d}x$$        
则称$X$为连续型随机变量。$f(x)$称为$x$的概率密度函数或密度函数，简称**概率密度**。   

**密度函数具有下面4个性质:**

$1.f(x)\geq 0$

$2.F(x)=\int_{-\infty}^{+\infty}{f(x)} \,{\rm d}x=1$

$3.对于任意实数x_1,x_2,(x_1<x_2),P(x_1\leq X\leq x_2)=F(x_2)-F(x_1)=\int_{x_1}{x_2}f(x)dx$

$4.若f(x)$在点$x$处连续，则有$F^{'}(x)= f(x)$  


**非离散型随机变量的特点:**

1.其可能取值不能一个一个地列举出来

2.非离散型随机变量取任一指定的实数值的概率都等于0

3.在计算连续型随机变量落在某一区间的概率时，可以不必区分该区间是开区间或闭区间或半闭区间

##### 离散与连续型随机变量的关系
$$P(X=x)\approx P(x<X\leq x+dx)\approx f(x)dx$$

## $\S2.2$ 分布函数
设$X$为随机变量，是$x$任意实数，则函数$F(x)=P(X\leq x)$称为随机变量X的分布函数，本质上是一个累积函数。  

$P(a<X\leq b)=F(b)-F(a)$  可以得到X落入区间(a,b]的概率。分布函数F(x)表示随机变量落入区间$(-\infty，x]$内的概率。  

**分布函数具有如下性质：**：  

1.$0\leq F(x)\leq 1,-\infty<x<+\infty$

2.$F(x)$是单调不减的函数，即$x_1<x_2$时，有$F(x_1)\leq F(x_2)$

3.$F(+\infty)=\lim_{x \to +\infty}F(x)=1$

4.$F(x+0)=\lim_{\Delta x \to -\infty }F(x+\Delta x)=F(x)$,&emsp;$F(x)$即是右连续的 

5.$P(a<X\leq b)=p(X\leq b)-p(x\leq a)=F(b)-F(a)$    

6.$P(X>a)=1-P(X\leq a)=1-F(a)$

7.$F(x)=\sum_{x_k\leq x}^{}P_k$

8.$F(x)=\int_{-\infty}^x f(x), F^{'}(x)=f(x)$

##### (1)0-1分布	
| 随机变量X | 0 | 1 |
| ---- | -- | :- | 
| 概率P&nbsp;&nbsp;&nbsp; | $p$ | $q$ | 

$$P(X=1)=p, P(X=0)=q$$

##### (2)二项分布
在$n$重贝努里试验中，设事件$A$发生的概率为$P$。事件$A$发生的次数是随机变量，设为$X$，则可能取值为$0,1,2,\ldots,n$。  

二项分布的分布律为：$$P(X=k)=P_n(k)={C_n}^kp^kq^(n-k)$$  
其中$q=1-p,0<p<1,k=0,1,2,\ldots,n$，则称随机变量$X$服从参数为$n$的二项分布,记为$X\sim B(n,p)$。  

当$n=1$时$P(X=k)=p^kq^(1-k)$，这就是（0-1）分布，所以（0-1）分布是二项分布的特例。

##### (3)泊松分布
设随机变量的分布律为:
$$P(X=k)=\frac{\lambda ^k}{k!}e^{-\lambda},\lambda>0,k=0,1,2\ldots$$  

则称随机变量$X$服从参数为$\lambda$的泊松分布，记为$X\sim \pi(\lambda)$或者$P(\lambda)$。  

泊松分布为二项分布的极限分布$(np=\lambda,n\to \infty)$。

##### (4)超几何分布	
$$P(X=k)=\frac{C_M^k \cdot C_{N-M}^{n-k}}{C_N^n},k=0,1,2\ldots,l；l=min(M,n)$$  

随机变量X服从参数为$n,N,M$的超几何分布，记为$X\sim H(n,N,M)$。

##### (5)几何分布
$$P(X=k)=q^{k-1}p,k=1,2,3,\ldots其中p\geq 0，q=1-p$$  

随机变量$X$服从参数为$p$的几何分布，记为$G(p)$。

##### (6)均匀分布	
设随机变量$X$的值只落在$[a,b]$内，其密度函数$f(x)$在$[a,b]$上为常数$\frac{1}{b-a}$，即：

$$f(x)=\begin{cases}
\frac{1}{b-a},& a\leq x\leq b \\ 
0,& 其它
\end{cases}$$  

则称$X$随机变量在$[a,b]$上服从均匀分布，记为$X\sim U(a，b)$。  

**分布函数**为：
$$F(x)=\int_{-\infty}^{x}f(x)d(x)=\begin{cases}
0,& x<a\\
\frac{x-a}{b-a},& a\leq x\leq b\\
1,& x>b
\end{cases}$$
                                        
当$a\leq x_1<x_2\leq b $时，X落在区间$(x_1,x_2)$内的概率为$P(x_1<X<x_2)=\frac{x_2-x_1}{b-a}$  

该随机变量X落在区间（a,b)中任意等长度的子区间内的可能性是相同的，或者说它落在（a,b)的子区间内的概率只依赖与子区间的长度而与子区间的位置无关。


##### (7)指数分布
$$f(x)=\begin{cases}
\lambda e^{-\lambda x},&x\geq 0\\
0,&x<0 
\end{cases}$$ 

其中$\lambda$,则称随机变量$X$服从参数为$\lambda$的指数分布。  

X的**分布函数**为
$$f(x)=\begin{cases}
1-e^{-\lambda x},&x>0 \\
0,&x<0
\end{cases}$$  
记住积分公式：
$$\int_{0}^{+\infty}x^ne^{-x}dx=n!$$  

服从指数分布的随机变量X具有以下性质：（无记忆性）  

对于任意$s,t>0$,有$P(X>s+t\;|\;X>s)=P(X>t)$  

证明:   
$$P(X>s+t\;|\;X>s)=\frac{P{(X>s+t)\bigcap 
(X>s)}}{p(X>s)}=\frac{P(X>s+t)}{p(X>s)}=\frac{1-F(s+t)}{1-F(s)}=\frac{e^{-(s+t)/\theta}}{e^{-s/\theta}}=e^{-t/\theta}=1-F(t)=1-P(X\leq t)=p(X>t)$$
       

##### (8)正态分布
设随机变量$X$的**密度函数**为：
$$f(x)=\frac{1}{\sqrt{2\pi\sigma}}e^{-\frac{(x-\mu)^2}{2\sigma^2}},-\infty<x<+\infty$$  

其中$\mu,\sigma>0$为常数，则称随机变量$X$服从参数$\mu$,$\sigma$的正态分布或高斯（Gauss）分布，记为$X\sim N(\mu,\sigma^2)$。  

$f(x)$**具有如下性质**：  

1.$f(x)$的图形是关于$x=\mu$对称的,对于任意$h>0$有$P\mu-h<X\leq\mu)=P(\mu<X\leq\mu+h)$

2.当$x=\mu$时，$f(\mu)=\frac{1}{\sqrt{2\pi\sigma}}$为最大值。  

$X$离$\mu$越远，$f(x)$的值越小，这表明对于同样长度的区间，当区间离$\mu$越远，$X$落在这个区间上的概率越小
   
$f(x)$**的形状特性**：  

1.固定$\sigma$，改变$\mu$的值，则图形沿着$Ox$轴平移，而不改变其形状，可见正态分布函数的概率密度曲线$y=f(x)$的位置完全由参数$\mu$所确定，称$\mu$为位置参数;  
 
2.固定$\mu$，改变$\sigma$的值，由于最大值,可知当$\sigma$越小图形变得越尖，因而$X$落在$\mu$附近概率越大;   
   
   
若$X~N(\mu,\sigma^2)$则$X$的**分布函数**为：
$$F(x)=\frac{1}{\sqrt{2\pi\sigma}}\int_{-\infty}^{x}e^{-\frac{(t-\mu)^2}{2\sigma^2}}dt$$  

参数$\mu=0,\sigma=1$时的正态分布称为标准正态分布，记为$X~N(0,1)$。   

**其密度函数记为**：
$$\phi(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}},-\infty <x< +\infty$$

**分布函数为**：
$$\Phi(x)=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{x}e^{-\frac{t^2}{2}}dt$$


$\Phi(x)$是不可求积函数，其函数值，已编制成表可供查用。
$\phi(x)$**具有如下性质**：  

1.$\Phi(1-x)=1-\Phi(x)$且$\Phi(0)=\frac{1}{2}$；  

2.如果$X~N(\mu,\sigma^2)$,则$\frac{X-\mu}{\sigma}~N(0,1)$；  

3.$P(x_1<X\leq x_2)=\Phi(\frac{x_2-\mu}{\sigma})-\Phi(\frac{x_1-\mu}{\sigma})$。

##### 分位数介绍	
下分位表：$P(X\leq \mu_\alpha)=\alpha$；  

上分位表：$P(X\geq \mu_\alpha)=\alpha$。

##### 超几何、二项、泊松分布之间的近似关系
定理(泊松定理) 在$n$重伯努利试验中, 事件$A$在每次试验中发生的概率为$p_n$(注意这与试验次数$n$有关), 如果$n\to \infty$时,$np_n \to \lambda$($\lambda$>0为常数).则对于任意给定的k,有
$$\lim_{n \to \infty}B(k,n,p_n)=lim_{n \to \infty}C_n^kP_n^k(1-p_n)^{n-k}=\frac{{\lambda}^k}{k!}e^{-\lambda}$$  

定理 :超几何分布的极限分布是二项分布  
即:在超几何分布中对于固定的$n,k$，如果
                                $$lim_{N \to +\infty}\frac{M}{N}=p $$
则有极限关系：
          $$lim_{N \to +\infty}\frac{C_M^kC_{N-M}^{n-k}}{C_N^n}=C_n^kp^k(1-p)^n-k $$

对所有的$0\leq k\leq n$都成立。

**3.三角分布**

三角分布，又称辛普森分布。是在样本量极少，但比较确定这个数据集中的最大值、最小值、众数时使用的一种估计。在概率论与统计学中，三角形分布是低限为a、众数为c、上限为b的连续概率分布。

在概率论与统计学中，三角形分布是低限为a、众数为c、上限为b的连续概率分布$$f(x|a,b,c)=\begin{cases}
\frac{2(x-a)}{(b-a)(c-a)}&a\leq x\leq c\\
\frac{2(b-x)}{(b-a)(b-c)}&c<x\leq b
\end{cases}$$

当c=a或者c=b,分布局可以进一步简化。例如，a=0,b=c=1,对于$0\leq x\leq1$ $$f(x)=2x$$ $$F(x)=x^2$$ $$E(X)=\frac{2}{3}$$ $$Var(X)=\frac{1}{18}$$

三角形分布通常用于表述只有优先采样数据的人口信息，尤其是已知变量之间的关系但是由于数据的收集成本太高而缺少采样数据的场合。 这通常是根据已知最小值与最大值从而推算合理的常见值。三角形分布经常用于商务决策，尤其是计算机模拟领域。通常，如果对结果的概率分布所知信息很少，例如仅仅知道最大值与最小值，那么可以使用平均分布模型。但是，如果已经知道了最可能出现的结果，那么就可以用三角形分布进行模拟。

## $\S3.1$联合概率分布
**定义：**
设（X,Y)是二维随机变量，对于任意实数x,y,二元函数：$F(x,y) = P{(X \leq x)\cap(Y \leq y)} = P(X \leq x,Y \leq y)$称为二维随机变量（X,Y）的分布函数或**联合分布函数**.
（X,Y）的联合分布函数F(x,y)具有以下性质：    
1)<center>$0 \leq F(x,y) \leq 1$ </center>

2)<center>$F( +\infty ,+\infty) = \lim\limits_{\substack{x\to+\infty\\ y\to+\infty}} F(x,y) = 1$,   
$F( -\infty ,-\infty) = \lim\limits_{\substack{x\to-\infty\\ y\to-\infty}} F(x,y) = 0$,   
$F( -\infty ,y) = \lim\limits_{x\to-\infty} F(x,y) = 0$,   
$F(x, -\infty) = \lim\limits_{y\to-\infty} F(x,y) = 0$</center>

3)F(x,y)分别对x和y是单调非减的，即 
<center>
当$x_1<x_2,有 F(x_1,y) \leq F(x_2,y)$,    
当$y_1<y_2,有 F(x,y_1) \leq F(x,y_2)$   </center>

4)F(x,y)分别是x和y的右连续函数，即   
$$\lim\limits_{x\to x_0+0}F(x,y_0) = F(x_0+0，y_0) =F(x_0,y_0)$$,   
$$\lim\limits_{y\to y_0+0}F(x_0,y) = F(x_0，y_0+0) =F(x_0,y_0)$$   

5)对任意的$a_1<b_1$,$a_2<b_2$,有$$F(b_1,b_2)-F(a_1,b_2)-F(a_2,b_1)+F(a_1,a_2) \geq 0$$     

**定义：**   
如果二维随机变量(X,Y)所有可能取值为$R^2$中有限或之多可列(x,y)个点时，称(X,Y)为**二维离散型随机变量**。二维离散型随机变量(X,Y)的**联合概率分布**(或概率分布、分布律、联合分布律)，表示:
$P(X=a_i,Y=b_j)=p_{a_i,b_j}$  

相应的(X,Y)的联合分布函数为:$$F(x,y) = \sum_{b_j \leq y}\sum_{a_i \leq x}p_{ij}$$    
它是一个取值位于[0,1]上关于x和y非减的二元阶梯函数，在$(a_i,b_j)$处有跳跃$p_{ij} = F(a_i,b_j)F(a_i,b_j-0) + F(a_i-0,b_j-0)$对于$R^2$上任一集合$B$,有$$P((X,Y) \in \mathrm{B}) = \sum_{(a_i,b_j) \in B}\sum p_{ij}$$

**定义：**   
如果二维随机变量(X,Y)的联合分布函数F(x,y)可表示成$$F(x,y) = \int_{-\infty}^y\int_{-\infty}^x f(u,v)\mathrm{d}u\mathrm{d}v$$ $$(-\infty < x < \infty, -\infty < y < \infty)$$
则称(X,Y)是**二维连续型随机变量**，而二元函数$f(x,y)$称为$(X,Y)$的**联合密度函数**，$f(x,y)$必须满足：   
1) $f(x,y) \geq 0$   
2) $\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}f(x,y)\mathrm{d}x\mathrm{d}y = 1$

对于$R^2$上任一集合，有：$$P((X,Y) \in \mathrm{B}) = \iint_\mathrm{B}f(x,y)\mathrm{d}x\mathrm{d}y$$   

二维连续型随机变量具有下列性质：   
1) (X,Y)的分布函数$F(x,y)$是(x,y)的二元连续函数；   
2）在$f(x,y)$的连续点处，有$$\frac{\partial^2}{\partial x \partial y}F(x,y) = f(x,y)$$；   
3) 对于$R^2$上任一条平面曲线$L$, 有$$P((X,Y) \in L) = 0$$   

**常用的二维连续型随机变量的分布：**   
1) 均匀分布   
如果(X,Y)有密度函数
\begin{equation}
f(x,y) = \begin{cases}
\frac{1}{G的面积}  &  (x,y) \in G \\
0  &  其他 \\
\end{cases}
\end{equation}

其中，$G$是$R^2$上的某一区域，则称(X,Y)服从$G$上的均匀分布。 

2） 二维正态分布   
如果(X,Y)有密度函数为
\begin{equation}
f(x,y) = \frac{1}{2\pi \sigma_1\sigma_2 \sqrt{1-\rho^2}}·\exp \{{-\frac{1}{2(1-\rho^2)}[\frac{(x-\mu_1)^2}{\sigma_1^2}-\frac{2\rho(x-\mu_1)(y-\mu_2)}{\sigma_1\sigma_2}+\frac{(y-\mu_2)^2}{\sigma_2^2}}]\}
\end{equation}

其中，$- \infty < \mu_1 , \mu_2 < \infty, -1 < \rho < 1, \sigma_1^2 > 0,\sigma_2^2 > 0 $,   
则称(X,Y)服从二维正态分布 $ N( \mu_1,\mu_2,\sigma_1^2,\sigma_2^2,\rho) $


## $\S3.2$边缘分布


**定义：**   
设$(X,Y)$的联合分布函数为$F(x,y)$,则称
\begin{equation}
 F_X(x) = P(X \leq x) = F(x,+\infty)
 \end{equation}
\begin{equation}
 F_Y(y) = P(Y \leq y) = F(+\infty,y)
 \end{equation}
 为$ X $ $ Y $的**边缘分布函数**。   
 
 设二维离散型随机变量$(X,Y)$的联合概率分布为$$ P(X=a_i,Y=b_j) = p_{ij}  $$
 $(i=1,2,...;j=1,2,...)$   
 则称
 \begin{equation}
 P(X=a_i) = p_i = \sum_{j=1}^{\infty}p_{ij}
 \end{equation}
 $(i=1,2,...)$ 
 和
 \begin{equation}
 P(Y=b_i) = p_i = \sum_{i=1}^{\infty}p_{ij}
 \end{equation}
 $(j=1,2,...)$    
 分别为$X$和$Y$的**边缘概率分布**。    
 
 设二维连续型随机变量$(X,Y)$有联合密度函数$f(x,y)$，则称
  \begin{equation}
f_X(x) = \int_{-\infty}^{\infty}f(x,y)\mathrm{d}y
 \end{equation}
 和
 \begin{equation}
f_Y(y) = \int_{-\infty}^{\infty}f(x,y)\mathrm{d}x
 \end{equation}
分别为$X$和$Y$的**边缘密度函数**


**定理：**    
当$(X,Y)~N(\mu_1,\mu_2,\sigma_1^2,\sigma_2^2,\rho)$时，$X$的边缘分布为$N(\mu_1,\sigma_1^2)$,$Y$的边缘分布为$N(\mu_2,\sigma_2^2)$.   
对于二维随机变量$(X,Y)$而言，当$Y$取定值$y$后，$X$仍是随机变量，我们称在$Y = y$的条件下$X$的分布为x的条件分布。   
在$Y=b_j$条件下$X$的条件概率分布为:
\begin{equation}
P{X = a_i | Y = b_j} = p_{ij}/p_{·j}\\
(i = 1,2,...)
\end{equation}

在$X = a_i$条件下 $Y$的条件概率分布为
\begin{equation}
P{Y = b_j | X = a_i} = p_{ij}/p_{i·}\\
(j = 1,2,...)
\end{equation}

设二维连续型随机变量$(X,Y)$有联合密度函数$f(x,y)$，则在$Y=y$条件下（其中$y$使得$f_Y(y)>0$）,X的条件密度函数为
\begin{equation}
f(X|Y)(x|y) = f(x,y)/f_Y(y)
\end{equation}

在$X=x$条件下（其中$x$使得$f_X(x)>0$）,$Y$的条件密度函数为
\begin{equation}
f(Y|X)(y|x) = f(x,y)/f_X(x)
\end{equation}

**定理：**   
当$(X,Y)~N(\mu_1,\mu_2,\sigma_1^2,\sigma_2^2,\rho)$时，   
在$Y=y$给定条件下:$X~N(\mu_1+\rho \frac{\sigma_1}{\sigma_2}(y-\mu_2),\sigma_1^2(1-\rho^2))$.   
在$X=x$给定条件下： $Y~N(\mu_2+\rho\frac{\sigma_2}{\sigma_1}(x-\mu_1),\sigma_2^2(1-\rho^2))$
 

## $\S$3.3 独立性 

**定义：**   
设$(X,Y)$是二维随机变量，如果$(X,Y)$的联合分布函数等于$X$和$Y$的边缘分布函数的乘积，即
\begin{equation}
F(x,y) = $F_X(x)F_Y(y)$\\
$(-\infty<x<\infty, -\infty<y<\infty)$
成立，称随机变量X和Y是**相互独立的**。    

**定理：**    
随机变量X和Y相互独立等价于对实数轴上任意两个集合A,B，有
$$P(X \in A,Y \in B) = P(X \in A)P(Y \in B)$$

**定理：**    
设随机变量$X$和$Y$相互独立，则$X$的任一函数$g(X)$与$Y$的任一函数$h(Y)$也相互独立。   

**定理：**    
设(X,Y)为二维离散型随机变量，联合概率分布为
$$P(X=a_i,Y=b_j)=p{ij}\\(i=1,2,...;j=1,2,...)$$   
则$X$,$Y$相互独立的充分必要条件是
<center>
$p_{ij} = p_i·p_{·j}$
</center>

**定理:**   
设(X,Y)为二维连续型随机变量，联合概率密度函数为$f(x,y)$，则$X$,$Y$
相互独立的充分必要条件是
$f(x,y) = f_X(x)·f_Y(y)$   
对$R$上一切$x,y$成立。其中，$f_X(x)= \int_{-\infty}^{\infty}f(x,y)\mathrm{d}y$,$f_Y(y)= \int_{-\infty}^{\infty}f(x,y)\mathrm{d}x$

**定理：**    
设(X,Y)~N(\mu_1,\mu_2,\sigma_1^2,\sigma_2^2,\rho),则$X$和$Y$相互独立的充分必要条件是$\rho = 0$

若不同的$(a_i,b_j)$有相同的$g(a_i,b_j)$,则相应的概率应相加合并。

\begin{align}
F_Z(z) &= P(g(X,Y) \leq z)\\

\end{align}

其中，$\mathrm{D}_z = {(x,y) \geq z}$.而$Z$的密度函数可通过$F_Z(z)$对$z$求导得到。当$X$,$Y$独立时，在公式(3.25)中只须略作变化，即把 $p_{ij}改为p_{i·}p(·j)$  (i=1,2,...;j=1,2,...),$f(x,y)$改为$f_X(x)f_Y(y)$ 即可。其中，$p_{i·} = P(X=a_i),p_{·j}=P(Y=b_j),f_X(x),f_Y(y)$分别为$X$,$Y$的密度函数

\begin{align}
f_Z(z) &= \int_{-\infty}^{\infty}f(x,z-x)\mathrm{d}x\\

\end{align}

设$X$,$Y$为独立的离散型分布，它们的概率分布分别为
$$P(X=i) = p_i\\P(Y=j) = q_j\\(i,j=1,2,...))$$

**定理（分布的可加性）：**    
1)设$X~B(m,p),Y~B(n,p)$且相互独立，则$Z=(X+Y)~B(m+n,p)$

2)设$X~P(\lambda_1),Y~P(\lambda_2)$且相互独立，则$Z=(X+Y)~P(\lambda_1+\lambda_2)$

e)$U = max(X,Y) 和 V=min(X,Y)$的分布   
设$X,Y$相互独立，$X$的分布函数为$F_X(x),Y$的分布函数为$F_Y(y)$,则    
$U = max(X,Y)$的分布函数为

\end{equation}
$ V=min(X,Y)$的分布函数为

F_V(v) = 1-(1-F_X(v))·(1-F_Y(v))
\end{equation}

更一般地,设$X_1,...,X_n$相互独立，且有相同的密度函数$f(x)$和分布函数$F(x)$，则   
$U = max(X_1,...,X_n)$的密度函数为   
\begin{equation}
f_U(u) = nF^{n-1}(u)·f(u)
\end{equation}

\begin{equation}
F_V(v) = n(1-F(v)^{n-1})·f(v)
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sy

loc,scale = 10, 1
s = np.random.logistic(loc,scale,10000)
count,bins,ignored = plt.hist(s, bins=50)

def logist(x, loc, scale):
    return sy.exp((loc-x)/scale)/(scale*(1+sy.exp((loc-x)/scale))**2)
plt.plot(bins, logist(bins, loc, scale)*count.max()/\
         logist(bins, loc,scale).max())
plt.title('Logistic Distribution')
plt.show()
```

## $\S$4.1 大数定律

在数学与统计学中，大数定律又称大数法则、大数律，是描述相当多次数重复实验的结果的定律。根据这个定律知道，样本数量越多，则其平均就越趋近**期望值**。

大数定律很重要，因为它“保证”了一些随机事件的均值的长期稳定性。人们发现，在重复试验中，随着试验次数的增加，事件发生的**频率**趋于一个稳定值；人们同时也发现，在对物理量的测量实践中，测定值的**算术平均**也具有稳定性。比如，我们向上抛一枚硬币，硬币落下后哪一面朝上是偶然的，但当我们上抛硬币的次数足够多后，达到上万次甚至几十万几百万次以后，我们就会发现，硬币每一面向上的次数约占总次数的二分之一，亦即偶然之中包含着必然。**切比雪夫定理的一个特殊情况。辛钦定理和伯努利大数定律都概括了这一现象，都称为大数定律。**


## $\S$4.2 切比雪夫大数定律

设随机变量 $X_1$,$X_2$, ... ,$X_n$, ... 相互独立（是指对于任意 n>1, $X_1$,$X_2$, ... ,$X_n$, ...是相互独立），且具有相同的数学期望和方差：$$E(X_k)=\mu, D(X_k)=\sigma^2(k=1,2,...)$$ 

作前n个随机变量的算术平均: $\overline{X}=\frac{1}{n} \sum\limits_{k=1}^n X_k $ ,

则对于任意正数 $\epsilon$ ,有
$$ \lim\limits_{n \to \infty}P\lbrace
|\overline{X}-\mu|<\epsilon \rbrace
= \lim \limits_{n \to \infty}P\lbrace
|\frac{1}{n}\sum\limits_{k=1}^n X_k-\mu|<\epsilon\rbrace
=1 $$

证明：

由于
$$ E[\frac{1}{n} \lim\limits_{k=1}^n X_k]=\frac{1}{n}\sum\limits_{k=1}^n E(X_k)=\frac{1}{n}*n\mu, $$ 

$$D[\frac{1}{n} \lim\limits_{k=1}^n X_k]=\frac{1}{n^2}\sum\limits_{k=1}^n D(X_k)=\frac{1}{n^2}*n\sigma^2=\frac{\sigma^2}{n},$$

由契比雪夫不等式可得
$$P\lbrace
{|\frac{1}{n}\sum\limits_{k=1}^n X_k-\mu|<\epsilon}\rbrace
\geq 1-\frac{\sigma^2/n}{\epsilon^2}$$

在上式中令$n \to \infty$, 并注意到概率不能大于1，即得
$$\lim \limits_{n \to \infty}P\lbrace
{|\frac{1}{n}\sum\limits_{k=1}^n X_k-\mu|<\epsilon} \rbrace
=1.$$

**契比雪夫不等式**

设随机变量X具有数学期望E(X)=μ，方差D(X)=$\sigma^2$,对于任意ε>0，都有
$$P\lbrace
|X-\mu|\geq \epsilon\rbrace
\leq \frac{\sigma^2}{\epsilon^2}
$$

方差越大，X落在区间外的概率越大，X的波动也就越大，与方差的意义统一了。等价公式
$$P\lbrace
|X-\mu|< \epsilon\rbrace
\geq 1- \frac{\sigma^2}{\epsilon^2}
$$
## $\S$4.3 伯努利大数定律

设$n_A$是n次独立重复试验中事件A发生的次数。p是事件A在每次试验中发生的概率，则对于任意正数$\epsilon$>0,有
$$\lim\limits_{n \to \infty}P\lbrace|\frac{n_A}{n}-p|<\epsilon\rbrace=1
或
\lim\limits_{n \to \infty}P\lbrace|\frac{n_A}{n}-p|\geq\epsilon\rbrace=0$$

证明：

因为$n_A~b(n,p)$,有
$$ n_A=X_1+X_2+ ... +X_n,$$

其中，$X_1,X_2, ... ,X_n, ...$相互独立，且都服从以p为参数的（0-1）分布，因而

$$E(X_k)=p,D(X_k)=p(1-p)(k=1,2, ... ,n),$$
由定理一得:
$$\lim\limits_{n \to \infty}P\lbrace|\frac{1}{n}(X_1+X_2+ ... +X_n)-p|<\epsilon\rbrace=1,$$

即
$$\lim\limits_{n \to \infty}P\lbrace|\frac{n_A}{n}-p|<\epsilon\rbrace=1$$



这个定理表明事件发生的频率的稳定性.

## $\S$4.4 辛钦大数定律

设随机变量$X_1,X_2, ... ,X_n$, ...相互独立，服从同一分布，且具有数学期望 $E(X_k)=\mu(k=1,2, ...)$,则对于任意正数$\epsilon$,有
$$\lim\limits_{n \to \infty}P\lbrace
|\frac{1}{n}\sum\limits_{k=1}^n X_k-\mu|<\epsilon
\rbrace=1$$

显然，伯努利大数定理是辛钦定理的特殊情况。

|辛钦大数定律|相互独立且同分布|存在| |估算期望|

```python
#Python模拟抛硬币辅助说明大数定律
#由于random.random()生成的数可以看做是服从区间[0,1)上的均匀分布，所以以0.5为界限，随机生成的数中大于0.5或小于0.5的概率应该是相同的(相当于硬币是均匀的)。这样就用随机数模拟出了实际的抛硬币试验。理论上试验次数越多(即抛硬币的次数越多)，正反面出现的次数之比越接近于1(也就是说正反面各占一半).
import random
import matplotlib.pyplot as plt


def flip_plot(minExp, maxExp):
    """
    Assumes minExp and maxExp positive integers; minExp < maxExp
    Plots results of 2**minExp to 2**maxExp coin flips
    """
    # 两个参数的含义，抛硬币的次数为2的minExp次方到2的maxExp次方，也就是一共做了(2**maxExp - 2**minExp)批次实验，每批次重复抛硬币2**n次

    ratios = []
    xAxis = []
    for exp in range(minExp, maxExp + 1):
        xAxis.append(2**exp)
    for numFlips in xAxis:
        numHeads = 0 # 初始化，硬币正面朝上的计数为0
        for n in range(numFlips):
            if random.random() < 0.5:  # random.random()从[0, 1)随机的取出一个数
                numHeads += 1  # 当随机取出的数小于0.5时，正面朝上的计数加1
        numTails = numFlips - numHeads  # 得到本次试验中反面朝上的次数
        ratios.append(numHeads/float(numTails))  #正反面计数的比值
    plt.title('Heads/Tails Ratios')
    plt.xlabel('Number of Flips')
    plt.ylabel('Heads/Tails')
    plt.plot(xAxis, ratios)
    plt.hlines(1, 0, xAxis[-1], linestyles='dashed', colors='r')
    plt.show()

flip_plot(4, 16)
```

## $\S$4.5 中心极限定理
中心极限定理是概率论中的一组定理。中心极限定理说明，在适当的条件下，大量相互独立随机变量的均值经适当标准化后依分布收敛于正态分布。这组定理是数理统计学和误差分析的理论基础，指出了大量随机变量之和近似服从正态分布的条件。

```python
import numpy as np
import matplotlib.pylab as plb
from scipy import stats
import math

number=[30,100,300,1000,5000,30000]
y=[]
for i in range(len(number)):
    ave_uniform=[]
    for j in range(1000):
        data=np.random.uniform(-1,1,number[i])
        variance=(1.0/3.0)/(1.0*number[i])
        summer=sum(data)
        ave_uniform.append(summer/(1.0*number[i]))
        Range=np.arange(-0.5,0.5,0.001)
    plb.subplot(230+i+1)
    plb.plot(Range,stats.norm.pdf(Range,0,math.sqrt(variance)))
    plb.hist(ave_uniform,bins=100)
plb.show()
```

这里将不同样本数下正态分布的均值求平均，每种样本数下重复1000次画线，与一个满足同样参数的正态分布样本集作比较，结果很明显。这里用正态分布是因为它的期望和方差很好手算，在用pdf函数的时候可以直接带入

**独立同分布的中心极限定理**

设随机变量$X_1,X_2, ... X_n, ...$相互独立，服从同一分布，且具有数学期望和方差：

$$E(X_k)=\mu,D(X_k)=\sigma^2(k=1,2, ...),$$

则随机变量之和$\sum\limits_{k=1}^nX_k$的标准化变量：

$$
Y_n=\frac{\sum\limits_{k=1}^n X_k-E(\sum\limits_{k=1}^n X_k)}
{\sqrt{D(\sum\limits_{k=1}^n X_k)}}
=\frac{\sum\limits_{k=1}^nX_k-n\mu}
{\sqrt{n}\sigma}
$$

分布函数$F_n(x)$对于任意x满足：
$$\lim\limits_{n\to\infty}F_n(x)=
\lim\limits_{n\to\infty}P\lbrace\frac
{\sum\limits_{k=1}^nX_k-n\mu}
{\sqrt{n}\sigma}
\leq x\rbrace=
\int_{-\infty}^x {\frac{1}{\sqrt{2\pi}}e^{-t^2/2}} {d}t
=\Phi(x)
$$

均值为μ，方差为$\sigma^2$>0的独立同分布的随机变量之和$\sum\limits_{k=1}^nX_k$的标准化变量，当n充分大时，有
$\frac
{\sum\limits_{k=1}^nX_k-\mu}
{\sigma/\sqrt{n}}
$~N(0,1),将式子左端改写成
$$\frac
{\sum\limits_{k=1}^nX_k-n\mu}
{\sigma/\sqrt{n}}
=\frac
{\overline X-\mu}
{\sigma/\sqrt{n}}
$$

当n充分大时，
$\frac
{\overline X-\mu}
{\sigma/\sqrt{n}}
$~N(0,1)或$\overline X$~N($\mu,\sigma^2/n$).这也就是说，均值为μ，方差为$\sigma^2>0$的独立同分布的随机变量$X_1,X_2, ... ,X_n,$的算术平均$\overline X=\frac{1}{n}\sum\limits_{k=1}^nX_k$,当n充分大时近似地服从均值为μ，方差为$\frac{\sigma^2}{n}$的正态分布。

**用中心极限定理生成正态分布，其本质就是生成n个独立同分布的均匀分布**

```python
import matplotlib.pyplot as plt
import numpy as np
def getNormal(SampleSize,n):
    result = []
    for i in range(SampleSize):
        # 利用中心极限定理，[0,1)均匀分布期望为0.5，方差为1/12
        iid = (np.mean(np.random.uniform(0,1,n))-0.5)*np.sqrt(12*n)
        result.append(iid)
    return result
# 生成10000个数，观察它们的分布情况
SampleSize = 10000
# 观察n选不同值时，对最终结果的影响
N = [1,2,10,1000] 
plt.figure(figsize=(10,20))
subi = 220
for index,n in enumerate(N):
    subi += 1
    plt.subplot(subi)
    normal = getNormal(SampleSize, n)
    # 绘制直方图
    plt.hist(normal,np.linspace(-4,4,81),facecolor="green",label="n={0}".format(n))
    plt.ylim([0,450])
    plt.legend()
    plt.show()
```

可以看到，n=1时其实就是均匀分布，随着 n逐渐增大，直方图轮廓越来越接近正态分布了。

**$\chi^2$分布**

设$x_{1},x_{2},...,x_{n}$为来自正态总体$N(\mu,\sigma^2)$的一个样本，则样本函数$$\omega=\frac{(n-1)S^2}{\sigma^2}\sim \chi^2(n-1)$$其中$\chi^2(n-1)$表示自由度为n-1的$\chi^2$分布。χ2检验是以χ2分布为基础的一种假设检验方法，主要用于分类变量。其基本思想是根据样本数据推断总体的分布与期望分布是否有显著性差异，或者推断两个分类变量是否相关或者独立。 一般可以设原假设为 H0：观察频数与期望频数没有差异，或者两个变量相互独立不相关。实际应用中，我们先假设H0成立，计算出χ2的值，χ2表示观察值与理论值之间的偏离程度。根据χ2分布，χ2统计量以及自由度，可以确定在H0成立的情况下获得当前统计量以及更极端情况的概率p。如果p很小，说明观察值与理论值的偏离程度大，应该拒绝原假设。否则不能拒绝原假设。A为实际值，T为理论值。
$$\chi^2=\sum{(A-T)^2}{T}$$

```python
#使用卡方检验：“豌豆黄圆：黄皱：绿圆:绿皱=315：101：108：32 是否符合9:3:3:1的理论比例”
import scipy.stats as stats
import pandas as pd
import numpy as np

observe=[315,101,108,32]
proportion=[9,3,3,1]
expect=[sum(observe)*i/sum(proportion) for i in proportion]
chisquare_val,p_val=stats.chisquare(observe,expect)
if p_val>0.05:
    result='Yes'
else:
    result='No'
result_table=pd.DataFrame({'chisquare_value':[chisquare_val],
    'p_value':[p_val],'result':[result]})
result_table
```
